```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('GTEx_Analysis_2017-06-05_v8_RNASeQCv1.1.9_gene_reads.gct', skiprows=2, sep='\t' )
attr = pd.read_csv("GTEx_Analysis_v8_Annotations_SampleAttributesDS.txt", sep='\t')
```

```python
print('original shape: ', all_data.shape)
```

```python
all_data.index = all_data.Name
all_data.drop(columns=['Name','Description'], inplace=True)
```

```python
non_expressed_gene = (all_data < 10).astype(int).sum(axis=1) #count how many samples with less than 5 reads
all_data = all_data[non_expressed_gene < all_data.shape[1]*0.75]

#save all gene ID for later use
np.save('/content/drive/MyDrive/muscle_promoter/training_data/selected_index.npy', all_data.index)

print('shape of trimmed dataframe: ', all_data.shape)
```

```python
all_data = all_data.div(all_data.sum(axis=0), axis=1)*1000000 #count per million
all_data.head(2)
```

# set a subset of blood sample aside

```python
!cp /content/drive/MyDrive/muscle_promoter/utils.py .
from utils import *
```

```python
attr = pd.read_csv("GTEx_Analysis_v8_Annotations_SampleAttributesDS.txt", sep='\t')
attr_blood = attr[attr.SMTSD=='Whole Blood']['SAMPID'].tolist()
attr_heart = attr[attr.SMTSD=='Heart - Left Ventricle']['SAMPID'].tolist()

attr_blood = intersection(all_data.columns.tolist(), attr_blood)
attr_heart = intersection(all_data.columns.tolist(), attr_heart)

len(attr_blood), len(attr_heart)
```

```python
patient_blood = remove_duplicate_in_list([patient_id(sample) for sample in attr_blood])
patient_heart = remove_duplicate_in_list([patient_id(sample) for sample in attr_heart])

common_patient = intersection(patient_blood, patient_heart)
len(patient_blood), len(patient_heart), len(common_patient)
```

```python
attr_blood_selected = []
for sample in attr_blood:
  if patient_id(sample) in common_patient:
    attr_blood_selected.append(sample)

attr_heart_selected = []
for sample in attr_heart:
  if patient_id(sample) in common_patient:
    attr_heart_selected.append(sample)

len(attr_blood_selected), len(attr_heart_selected)
```

```python
#rearrange attr_blood and _heart according to the list of common_patient

attr_blood_selected_order = []
attr_heart_selected_order = []

for patient in common_patient:
  for blood_sample in attr_blood_selected:
    if patient_id(blood_sample) == patient: attr_blood_selected_order.append(blood_sample)
  for heart_sample in attr_heart_selected:
    if patient_id(heart_sample) == patient: attr_heart_selected_order.append(heart_sample)

for i in range(len(common_patient)):
  assert common_patient[i] == patient_id(attr_blood_selected_order[i]) == patient_id(attr_heart_selected_order[i])

print(len(common_patient), len(attr_blood_selected_order), len(attr_heart_selected_order))
```

```python
blood = all_data[attr_blood_selected_order]
heart = all_data[attr_heart_selected_order]
```

```python
#save all blood and heart sample for later analysis

blood.to_csv('/content/drive/MyDrive/muscle_promoter/training_data/cpm_blood_366sampleordered.txt', sep='\t', index=True)
heart.to_csv('/content/drive/MyDrive/muscle_promoter/training_data/cpm_heart_366sampleordered.txt', sep='\t', index=True)
```

```python
from random import sample

random_indexes = sample(range(len(common_patient)), k=75)
attr_blood_selected_order_testset = [attr_blood_selected_order[i] for i in random_indexes]
attr_heart_selected_order_testset = [attr_heart_selected_order[i] for i in random_indexes]
```

```python
blood_testset = blood[attr_blood_selected_order_testset]
heart_testset = heart[attr_heart_selected_order_testset]

blood_testset.to_csv('/content/drive/MyDrive/muscle_promoter/training_data/cpm_blood_testset.txt', sep='\t', index=True)
heart_testset.to_csv('/content/drive/MyDrive/muscle_promoter/training_data/cpm_heart_testset.txt', sep='\t', index=True)
```

```python
blood_trainset = blood.drop(columns=attr_blood_selected_order_testset)
heart_trainset = heart.drop(columns=attr_heart_selected_order_testset)

blood_trainset.to_csv('/content/drive/MyDrive/muscle_promoter/training_data/cpm_blood_trainset.txt', sep='\t', index=True)
heart_trainset.to_csv('/content/drive/MyDrive/muscle_promoter/training_data/cpm_heart_trainset.txt', sep='\t', index=True)
```

```python
#drop all sample related to testset to not have leakage in data during training

all_data = all_data.drop(columns=attr_blood_selected_order_testset)
all_data = all_data.drop(columns=attr_heart_selected_order_testset)
all_data.shape
```
